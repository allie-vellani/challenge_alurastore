# Desafio Alura Store – Análise de Desempenho das Lojas

O objetivo deste projeto é analisar o desempenho das quatro lojas da rede fictícia Alura Store e recomendar qual delas o Sr. João deve vender. A análise considera métricas como faturamento, ticket médio, avaliação dos clientes, desempenho por categoria de produto e localização geográfica das vendas.


```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)
```

```python
import matplotlib.pyplot as plt

# Análise exploratória dos datasets
def explorar_dataset(df, nome='Dataset'):
    print(f"===== {nome} =====\n")

    print(" Dimensões:", df.shape)
    print("\n Primeiras linhas:")
    display(df.head())

    print("\n Tipos de dados:")
    print(df.dtypes)

    print("\n Valores únicos por coluna:")
    print(df.nunique())

    print("\n Duplicados:", df.duplicated().sum())
    print(" Valores nulos:", df.isnull().sum().sum())

explorar_dataset(loja, 'Loja 1')
explorar_dataset(loja2, 'Loja 2')
explorar_dataset(loja3, 'Loja 3')
explorar_dataset(loja4, 'Loja 4')
```



---



#1. Análise do faturamento





```python
# Função para calcular o faturamento total de uma loja
def calcular_faturamento(df, nome_dataset='') -> float:
    # Soma todos os valores da coluna 'Preço' para calcular o total vendido
    total = df['Preço'].sum()

    # Mostra o resultado no console, com duas casas decimais
    print(f" Faturamento total para {nome_dataset}: R$ {round(total, 2)}")

    # Retorna o valor total
    return total

# Lista com os DataFrames de cada loja
datasets = [loja, loja2, loja3, loja4]

# Lista com os nomes das lojas (na mesma ordem dos DataFrames)
nomes = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

# Dicionário para armazenar o faturamento total de cada loja
faturamentos = {}

# Loop para calcular e armazenar o faturamento de cada loja
for df, nome in zip(datasets, nomes):
    faturamentos[nome] = calcular_faturamento(df, nome)
```



---



# 2. Vendas por Categoria


```python
# Função para calcular a quantidade de vendas por categoria de produto
def vendas_por_categoria(df, nome_dataset=''):
    # Agrupa os dados pela coluna 'Categoria do Produto'
    # Conta quantas vezes cada categoria aparece (ou seja, quantos produtos foram vendidos em cada uma)
    vendas = df.groupby('Categoria do Produto')['Produto'].count().sort_values(ascending=False)

    # Exibe o nome da loja e o resultado
    print(f"\n Categorias mais vendidas - {nome_dataset}")
    display(vendas)

    # Retorna a série com os dados de vendas por categoria
    return vendas

# Dicionário para armazenar as vendas por categoria de cada loja
vendas = {}

# Loop para aplicar a função em todas as lojas
for df, nome in zip(datasets, nomes):
    vendas[nome] = vendas_por_categoria(df, nome)

# Converte o dicionário em um DataFrame para facilitar a visualização comparativa entre lojas
df_vendas = pd.DataFrame(vendas)
```



---



# 3. Média de Avaliação das Lojas

```python
# Função para calcular a média das avaliações dos clientes em uma loja
def media_avaliacoes(df, nome_dataset=''):
    # Calcula a média da coluna 'Avaliação da compra'
    media = df['Avaliação da compra'].mean()

    # Exibe o nome da loja e a média arredondada para 2 casas decimais
    print(f"Média de avaliações para {nome_dataset}: {round(media, 2)}")

    # Retorna o valor da média
    return media

# Dicionário para armazenar a média de avaliações de cada loja
media = {}

# Loop para calcular a média de avaliações em todas as lojas
# 'datasets' contém os DataFrames e 'nomes' contém os nomes das lojas
for df, nome in zip(datasets, nomes):
    media[nome] = media_avaliacoes(df, nome)
```



---



# 4. Produtos Mais e Menos Vendidos

```python
# Função para encontrar os 10 produtos mais vendidos em uma loja
def mais_vendidos(df, nome_dataset=''):
    # Agrupa os dados pelo nome do produto e conta quantas vezes cada um aparece
    max_vendas = df.groupby('Produto')['Produto'].count().sort_values(ascending=False)

    # Exibe o nome da loja e os 10 produtos mais vendidos
    print(f"\n Produtos mais vendidos - {nome_dataset}")
    display(max_vendas.head(10))

    # Retorna a lista completa
    return max_vendas

# Dicionário para armazenar os produtos mais vendidos de cada loja
mais_vendidos_por_loja = {}

# Loop para aplicar a função em todas as lojas
for df, nome in zip(datasets, nomes):
    mais_vendidos_por_loja[nome] = mais_vendidos(df, nome)


# Função para encontrar os 10 produtos menos vendidos em uma loja
def menos_vendidos(df, nome_dataset=''):
    # Mesmo processo da função acima, mas agora ordena do menor para o maior
    min_vendas = df.groupby('Produto')['Produto'].count().sort_values(ascending=True)

    # Exibe o nome da loja e os 10 produtos menos vendidos
    print(f"\n Produtos menos vendidos - {nome_dataset}")
    display(min_vendas.head(10))

    # Retorna a lista completa
    return min_vendas

# Dicionário para armazenar os produtos menos vendidos de cada loja
menos_vendidos_por_loja = {}

# Loop para aplicar a função em todas as lojas
for df, nome in zip(datasets, nomes):
    menos_vendidos_por_loja[nome] = menos_vendidos(df, nome)
```

# 5. Frete Médio por Loja


```python
# Função para calcular o frete médio de um DataFrame
def frete_medio(df, nome_dataset=''):
    # Calcula a média da coluna 'Frete'
    frete = df['Frete'].mean()

    # Exibe o resultado formatado com duas casas decimais
    print(f"Frete médio para {nome_dataset}: {frete:.2f}")

    # Retorna o valor do frete médio
    return frete

# Dicionário para armazenar o frete médio de cada loja
frete_medio_por_loja = {}

# Loop para calcular o frete médio de cada loja
for df, nome in zip(datasets, nomes):
    # Chama a função e armazena o resultado no dicionário
    frete_medio_por_loja[nome] = frete_medio(df, nome)
```

#**Relatório de Análise das Lojas**

## Introdução

O Sr. João, dono da rede Alura Store, está pronto para começar um novo projeto e decidiu vender uma de suas quatro lojas para levantar capital. Mas antes de tomar essa decisão importante, ele precisa saber: qual loja está com o desempenho mais fraco?

Para ajudar nessa escolha, realizei uma análise completa dos dados de vendas das quatro lojas. Olhei para o faturamento total, avaliações dos clientes, categorias e produtos mais vendidos, além do frete médio em cada unidade.

Usando ferramentas como Pandas e Matplotlib, conseguimos transformar os dados em insights visuais e objetivos, comparando o desempenho de cada loja. Ao final, trazemos uma recomendação clara: qual loja o Sr. João deve vender — e o porquê dessa escolha.

```python
# Criando gráfico de barras horizontais para mostrar o faturamento total por loja

# Define o tamanho da figura do gráfico
plt.figure(figsize=(8, 6))

# Cria listas com os nomes das lojas e os valores de faturamento
lojas = list(faturamentos.keys())
valores = list(faturamentos.values())

# Cria o gráfico de barras horizontal
# 'height' define a espessura das barras e 'color' define a cor
plt.barh(lojas, valores, height=0.5, color='skyblue')

# Adiciona um título ao gráfico
plt.title('Faturamento Total por Loja', fontsize=16, fontweight='bold')

# Adiciona o nome do eixo X
plt.xlabel('Faturamento (R$)', fontsize=12)

# Ajusta o layout para evitar que os textos fiquem cortados
plt.tight_layout()

# Exibe o gráfico
plt.show()
```

O gráfico mostra o quanto cada loja da Alura Store faturou no período analisado. Dá pra ver uma diferença considerável entre os desempenhos.

A Loja 1 lidera com folga, atingindo um faturamento total de R$ 1.534.509,12, se destacando como a mais lucrativa da rede. Logo depois vêm a Loja 2 (R\$ 1.488.459,06) e a Loja 3 (R\$ 1.464.025,03), com números bem próximos — o que mostra uma boa disputa entre elas.

Já a Loja 4 ficou bem abaixo das outras, com um total de R\$ 1.384.497,58. Esse resultado coloca a unidade como a de menor desempenho financeiro da rede.

```python
# Criando gráfico de barras verticais para mostrar a média de avaliações por loja

# Ordena as médias de avaliação do menor para o maior (para facilitar a leitura do gráfico)
media_ordenada = dict(sorted(media.items(), key=lambda item: item[1]))

# Separa os nomes das lojas e os valores das médias em listas
lojas = list(media_ordenada.keys())
avaliacoes = list(media_ordenada.values())

# Define o tamanho do gráfico
plt.figure(figsize=(8, 6))

# Cria o gráfico de barras verticais
plt.bar(lojas, avaliacoes, color='#6A5ACD', width=0.5)

# Adiciona os valores numéricos acima de cada barra, com duas casas decimais
for i, v in enumerate(avaliacoes):
    plt.text(i, v + 0.005, f"{v:.2f}", ha='center', fontname='DejaVu Sans', fontsize=10)

# Define o nome do eixo Y (avaliacoes)
plt.ylabel('Média de Avaliação', fontsize=12, fontname='DejaVu Sans')

# Adiciona um título ao gráfico
plt.title('Média de Avaliações por Loja', fontsize=16, fontweight='bold', fontname='DejaVu Sans')

# Define os limites do eixo Y para destacar melhor as diferenças entre as notas
plt.ylim(3.95, 4.1)

# Personaliza o tamanho e a fonte dos números dos eixos
plt.xticks(fontsize=11, fontname='DejaVu Sans')
plt.yticks(fontsize=10, fontname='DejaVu Sans')

# Define uma cor clara para o fundo do gráfico (opcional, só visual)
plt.gca().set_facecolor('#f9f9f9')

# Ajusta o layout para evitar cortes no gráfico
plt.tight_layout()

# Exibe o gráfico
plt.show()
```

Esse gráfico mostra a média das avaliações que cada loja da Alura Store recebeu dos clientes depois da compra. As notas vão de 1 a 5, sendo 5 a melhor possível.

A Loja 3 ficou no topo, com média 4.05, seguida de pertinho pela Loja 2, com 4.04. As duas mostram um bom nível de satisfação dos clientes — o que pode indicar entregas rápidas, atendimento bacana ou produtos de qualidade.

A Loja 4 teve uma média de 4.00, ficando ali no meio do caminho. Já a Loja 1 teve o pior desempenho nesse critério, com 3.98, sendo a única com nota abaixo de 4. Isso pode ser um sinal de que a experiência dos clientes por lá precisa de atenção.

```python
# Gerando gráficos de pizza para mostrar a distribuição de vendas por categoria em cada loja

# Loop para percorrer cada loja e sua respectiva série de vendas por categoria
for nome, serie in vendas.items():
    # Define o tamanho da figura do gráfico (largura x altura)
    plt.figure(figsize=(7, 7))

    # Cria o gráfico de pizza (pie chart)
    plt.pie(serie.values, labels=serie.index, autopct='%1.1f%%', startangle=140)

    # Adiciona um título com o nome da loja
    plt.title(f'Distribuição de Vendas por Categoria - {nome}')

    # Garante que o gráfico fique como um círculo perfeito
    plt.axis('equal')

    # Ajusta o layout para evitar cortes nos rótulos
    plt.tight_layout()

    # Exibe o gráfico
    plt.show()
```

Outras categorias como “brinquedos”, “eletrodomésticos” e “esporte e lazer” também têm presença forte. Já categorias como “instrumentos musicais”, “livros” e “utilidades domésticas” aparecem com menos frequência nas vendas — talvez por serem mais de nicho ou com menor rotatividade.

```python
# Trazendo dados existentes
avaliacoes = [media[nome] for nome in nomes]
faturamento = [faturamentos[nome] for nome in nomes]

# Criando o gráfico de dispersão
plt.figure(figsize=(8, 6))
plt.scatter(avaliacoes, faturamento, s=100, color='mediumseagreen')

# Adiciona os nomes das lojas nos pontos
for i, nome in enumerate(nomes):
    plt.text(avaliacoes[i] + 0.005, faturamento[i], nome, fontsize=10)

plt.xlabel('Média de Avaliação')
plt.ylabel('Faturamento (R$)')
plt.title('Avaliação x Faturamento por Loja')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Esse gráfico de dispersão junta duas informações super importantes: a média das avaliações dos clientes (eixo X) e o faturamento total de cada loja (eixo Y). A ideia aqui é visualizar, de forma clara, como cada loja está se saindo tanto em satisfação quanto em vendas.

As lojas que aparecem mais à direita e mais acima são as que combinam bons resultados financeiros com boas avaliações — ou seja, estão indo bem nos dois aspectos. Já as lojas mais à esquerda e abaixo do gráfico estão em uma posição mais delicada: baixo faturamento e menor avaliação, o que indica um desempenho mais fraco de forma geral.

Nesse cenário, a Loja 4 se destaca negativamente por estar no canto inferior esquerdo do gráfico.

```python
import folium

for df, nome in zip(datasets, nomes):
    df['Loja'] = nome

# Unindo todos os dados em um único DataFrame
df_total = pd.concat(datasets, ignore_index=True)

# Agrupar vendas por estado e por loja
vendas_por_estado_loja = df_total.groupby(['Local da compra', 'Loja']).size().unstack(fill_value=0)

# Selecionar os 5 estados com maior volume total de vendas
top_estados = vendas_por_estado_loja.sum(axis=1).sort_values(ascending=False).head(5).index

# Filtrar os dados para os estados selecionados
top_vendas_por_estado = vendas_por_estado_loja.loc[top_estados]

# Calcular faturamento por estado e por loja
faturamento_por_estado_loja = df_total.groupby(['Local da compra', 'Loja'])['Preço'].sum().unstack(fill_value=0)
faturamento_top_estados = faturamento_por_estado_loja.loc[top_estados]

# Calcular ticket médio (faturamento / quantidade de vendas)
ticket_medio_por_loja = faturamento_top_estados.div(top_vendas_por_estado)

#  Coordenadas aproximadas dos estados (top 5)
coordenadas_estados = {
    'SP': [-23.55, -46.63],
    'RJ': [-22.90, -43.20],
    'MG': [-18.10, -44.38],
    'RS': [-30.03, -51.23],
    'PR': [-25.43, -49.27],
}

# Cores para cada loja
cores_lojas = {
    'Loja 1': 'blue',
    'Loja 2': 'green',
    'Loja 3': 'purple',
    'Loja 4': 'red'
}

# Criação do Mapa com Ticket Médio por Loja e Estado

# Criar o mapa base
mapa_lojas = folium.Map(location=[-22.0, -50.0], zoom_start=5)

# Adiciona círculos por loja, em cada estado, usando cor e deslocamento leve
for estado in coordenadas_estados:
    lat, lon = coordenadas_estados[estado]

    for i, loja in enumerate(ticket_medio_por_loja.columns):
        deslocamento = (i - 1.5) * 0.5
            location=[lat, lon + deslocamento],
            radius=ticket_medio_por_loja.loc[estado, loja] / 30,
            color=cores_lojas[loja],
            fill=True,
            fill_color=cores_lojas[loja],
            fill_opacity=0.6,
            popup=f"{loja} - {estado}: R$ {ticket_medio_por_loja.loc[estado, loja]:.2f}"
        ).add_to(mapa_lojas)

# Exibe o mapa interativo
mapa_lojas
```

Esse mapa mostra a média de valor por venda (ticket médio) das quatro lojas da Alura Store nos cinco estados com maior volume de vendas: SP, RJ, MG, RS e PR.

Cada bolinha representa uma loja — com cores diferentes e posições levemente deslocadas para facilitar a visualização:O tamanho do círculo indica o ticket médio: quanto maior, mais alto o valor médio de cada venda naquele estado.

O que o mapa mostra é que, mesmo presente nos principais mercados, a Loja 4 (vermelha) tem círculos menores na maioria dos estados, ou seja, um ticket médio mais baixo comparado às outras unidades. Isso significa que ela vende, mas com menor valor por transação, o que prejudica seu faturamento total.

## Recomendação Final

Após a análise completa dos dados das quatro unidades da Alura Store, a recomendação é que o **Sr. João venda a Loja 4**.

### Por que a Loja 4?

Mesmo atuando fortemente em estados estratégicos como **São Paulo, Rio de Janeiro, Minas Gerais, Rio Grande do Sul** e **Paraná** — que concentram o maior volume de vendas da rede — a Loja 4 não consegue transformar essa presença em desempenho financeiro relevante.

### Principais pontos a respeito da Loja 4:

- **Menor faturamento total da rede.**

- **Ticket médio mais baixo** nos principais estados, mesmo com alto volume de vendas. Isso indica que as transações da Loja 4 têm, em média, menor valor, o que prejudica o resultado final.

- **Avaliação média regular (4.00).** Enquanto outras lojas conseguem equilibrar volume e satisfação, a Loja 4 não se sobressai em nenhum dos dois.

- **Atuação geográfica ineficiente:** os mapas mostram que a Loja 4 está presente em áreas de alto potencial de vendas, mas não se destaca em nenhum estado de forma significativa.

- **Ausência de destaque nas principais categorias.** Embora atue nos mesmos segmentos das outras lojas, a Loja 4 não se diferencia positivamente em nenhuma categoria relevante, o que enfraquece sua competitividade

## Conclusão

A Loja 4 não sofre por falta de mercado — ela está inserida em regiões promissoras — mas não consegue competir bem com as outras unidades, seja em faturamento, valorização por venda ou diferenciação.

Por isso, ela se mostra a **loja menos eficiente da rede** e a **melhor candidata para ser vendida**, permitindo que o Sr. João concentre seus recursos nas unidades com maior retorno e potencial de crescimento.
